```python
# Install/imports
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense


def build_lstm_model(vocab_size, embedding_dim=64, max_length=100, lstm_units=64, dropout_rate=0.5, num_classes=3):
    model = Sequential([
        Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
        Bidirectional(LSTM(lstm_units)),
        Dropout(dropout_rate),
        Dense(num_classes, activation='softmax')
    ])
    return model

# Example: build and print summary
if __name__ == '__main__':
    m = build_lstm_model(vocab_size=5000, embedding_dim=64, max_length=100, lstm_units=64, dropout_rate=0.5, num_classes=3)
    m.summary()
```
